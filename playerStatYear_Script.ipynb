```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
file1 = pd.read_csv('players.csv')
file2 = pd.read_csv('games.csv')
file3 = pd.merge(file1,file2, on='player_id')
print(file3.columns)
```

```python
def getYears(file,s, e):
    year = file[(file["year"] >= s) & (file["year"] <= e)]
    return year
```

```python
year = getYears(file3, 2016, 2017)
year.sort_values(by="name")
year.columns
```

```python
yearlystat = year.groupby(['name', 'position', 'height', 'weight', 'current_team','draft_round',
       'draft_position', 'draft_year','year','current_salary']).sum().reset_index()
yearlystat.columns
yearlystat
```

```python
df = DataFrame(yearlystat)
print(df)
df.to_excel(r'C:\Users\seneyd\Desktop\WIT Classes\Summer 2019\Senior Project\2016-2017.xlsx', index = None, header=True)
```
